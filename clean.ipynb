# Channel Volume

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Business_case/Data/Channel_Volume.csv")
df
```

```python
df = df.rename(columns = {"Volume": "Volume (millions of liters)"})
```

```python
mask = df["Unit"] == "000 litres"
df.loc[mask, "Volume (millions of liters)"] = df.loc[mask, "Volume (millions of liters)"].div(1000)
df
```

```python
df = df.drop(columns=["Industry", "Unit", "Edition", "Year_date"])
df
```

```python
print(df.isna().any())
```

```python
print(df[df["Category"].isna()])
```
